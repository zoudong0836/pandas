```python
import pandas as pd
```

```python
left = pd.DataFrame({
    'key' : ['K0', 'K1', 'K2', 'K3'],
    'A' : ['A0', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3']
})

right = pd.DataFrame({
    'key' : ['K0', 'K1', 'K2', 'K3'],
    'C' : ['C0', 'C1', 'C2', 'C3'],
    'D' : ['D0', 'D1', 'D2', 'D3']
})

print(left)
print(right)
```

```python
res = pd.merge(left, right, on='key')    # 基于'key'列进行合并
res
```

```python
 left = pd.DataFrame({
    'key1' : ['K0', 'K1', 'K2', 'K3'],
    'key2' : ['K0', 'K1', 'K0', 'K1'],
    'A' : ['A0', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3']
})

right = pd.DataFrame({
    'key1' : ['K0', 'K1', 'K2', 'K3'],
    'key2' : ['K0', 'K0', 'K0', 'K0'],
    'C' : ['C0', 'C1', 'C2', 'C3'],
    'D' : ['D0', 'D1', 'D2', 'D3']
})

print(left)
print(right)
```

```python
# how = ['left', 'right', 'inner', 'outer']
res = pd.merge(left, right, on=['key1', 'key2'], how='outer')  # how 默认值为'inner'
res
```

```python
res = pd.merge(left, right, on=['key1', 'key2'], how='outer', indicator=True) 
res
```

```python
res = pd.merge(left, right, on=['key1', 'key2'], how='inner')
res
```

```python
res = pd.merge(left, right, on=['key1', 'key2'], how='left')
res
```

```python
left = pd.DataFrame({
    'A' : ['A0', 'A1', 'A2'],
    'B' : ['B0', 'B1', 'B2']},
    index = ['K0', 'K1', 'K2']
)

right = pd.DataFrame({
    'C' : ['C0', 'C2', 'C3'],
    'D' : ['D0', 'D2', 'D3']},
    index = ['K0', 'K2', 'K3']
)

print(left)
print(right)
```

```python
res = pd.merge(left, right, left_index=True, right_index=True, how='outer')
res
```

```python
boys = pd.DataFrame({'k' : ['K0', 'K1', 'K2'], 'age' : [1, 2, 3]})
girls = pd.DataFrame({'k' : ['K0', 'K0', 'K3'], 'age' : [4, 5, 6]})
print(boys)
print(girls)
```

```python
res = pd.merge(boys, girls, on='k', suffixes=['_boy', '_girl'], how='outer')
res
```
