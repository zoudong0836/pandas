# 折线图

+ 折线图是用直线段将各数据连接起来组成的图形
+ 常用来观察数据随时间变化的趋势

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
date, open, close = np.loadtxt('data/000001.csv', delimiter=',', skiprows=1,
                               converters={
                                   # 把字符串格式是时间转换成数值 (numpy只接收数值)
                                   0: lambda x: mdates.date2num(datetime.strptime(x.decode(), "%m/%d/%Y"))
                               }, usecols=(0, 1, 4), unpack=True)
date
```

```python
plt.plot(date, open)
```

```python
# plot_date 能自动识别时间

plt.plot_date(date, open)
```

```python
plt.plot_date(date, open, linestyle='--', color='red', marker='o')
plt.plot_date(date, close, linestyle='-', color='green', marker='<')
```
