## 设置坐标轴

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 100)
y1 = 2*x + 1
y2 = x**2

plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='-')
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y1 = 2*x + 1
y2 = x**2

# 设置x, y范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))


plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='-')
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y1 = 2*x + 1
y2 = x**2

# 设置x, y范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# 设置x, y轴描述信息
plt.xlabel('I AM X')
plt.ylabel('I AM Y')


plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='-')
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y1 = 2*x + 1
y2 = x**2

# 设置x, y范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# 设置x, y轴描述信息
plt.xlabel('I AM X')
plt.ylabel('I AM Y')

# 设置x, y轴刻度
x_new_ticks = np.linspace(-2, 2, 11)
plt.xticks(x_new_ticks)
plt.yticks([-1, 0, 1, 2, 3], ['level1', 'level2', 'level3', 'level4', 'level5'])

plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='-')
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y1 = 2*x + 1
y2 = x**2

# 设置x, y范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# 设置x, y轴描述信息
# plt.xlabel('I AM X')
# plt.ylabel('I AM Y')

# 设置x, y轴刻度
x_new_ticks = np.linspace(-2, 2, 11)
plt.xticks(x_new_ticks)
plt.yticks([-1, 0, 1, 2, 3], ['level1', 'level2', 'level3', 'level4', 'level5'])

# gca 
ax = plt.gca()

# 把右边框和上边框去掉
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 把 x 轴的刻度设置为 ‘bottom’
ax.xaxis.set_ticks_position('bottom')
# 把 y 轴的刻度设置为 'left'
ax.yaxis.set_ticks_position('left')

# 设置 bottom 和 left 对应到 0 点
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='-')
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y1 = 2*x + 1
y2 = x**2

# 设置x, y范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# 设置x, y轴描述信息
plt.xlabel('I AM X')
plt.ylabel('I AM Y')

# 设置x, y轴刻度
x_new_ticks = np.linspace(-2, 2, 11)
plt.xticks(x_new_ticks)
plt.yticks([-1, 0, 1, 2, 3], ['level1', 'level2', 'level3', 'level4', 'level5'])

L1, = plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
L2, = plt.plot(x, y2, color='blue', linewidth=5.0, linestyle='-')

# 设置图例
plt.legend(handles=[L1, L2], labels=['test1', 'test2'], loc='best')

plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y1 = 2*x + 1

# 设置x, y范围
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# 设置x, y轴描述信息
# plt.xlabel('I AM X')
# plt.ylabel('I AM Y')


# gca
ax = plt.gca()

# 把右边框和上边框去掉
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 把 x 轴的刻度设置为 ‘bottom’
ax.xaxis.set_ticks_position('bottom')
# 把 y 轴的刻度设置为 'left'
ax.yaxis.set_ticks_position('left')

# 设置 bottom 和 left 对应到 0 点
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

plt.plot(x, y1, color='red', linewidth=1.0, linestyle='-')


# 选定直线上的一个点
x0 = 0.5
y0 = 2*x0 + 1

# 画点
plt.scatter(x0, y0, s=50, color='b')
# 画虚线
plt.plot([x0, x0], [y0, 0], 'k--', lw=2)

# 绘制标注
plt.annotate(r'$2x+1=%s$' % y0, xy=(x0, y0), xytext=(+30, -30), textcoords='offset points', 
             fontsize=16, arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))

plt.text(-1, 2, r'$this\ is\ the\ text$', fontdict={'size' : '16', 'color' : 'r'}) 

plt.show()
```

## 直方图

```python
x = np.arange(10)
y = 2**x + 10

plt.bar(x, y)
plt.show()
```

```python
x = np.arange(10)
y = 2**x + 10

plt.bar(x, -y)
plt.show()
```

```python
x = np.arange(10)
y = 2**x + 10

plt.bar(x, y, facecolor='#9999ff', edgecolor='white')
plt.show()
```

```python
x = np.arange(10)
y = 2**x + 10

plt.bar(x, y, facecolor='#9999ff', edgecolor='white')

for x, y in zip(x, y):
    plt.text(x-0.1, y, '%.2f' % y, ha='center', va='bottom')

plt.show()
```
