## jointplot 可以方便地展示两个变量之间的关系和边缘分布

```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as sci
import numpy as np
import pandas as pd

sns.set(style='white')
```

## 二维高斯分布

jointplot 共有5种类型：
+ 散点图 scatter (默认)
+ 六边形直方图 hex (hexagonal bins)
+ 回归 reg
+ 残差 resid
+ 核密度估计 kde

```python
x, y = np.random.multivariate_normal([0, 0], [(1, 0), (0, 1)], 1000).T
sns.jointplot(x, y)
```

```python
# 用六边形直方图取代中间的散点图

sns.jointplot(x, y, kind='hex')
```

```python
# 回归

sns.jointplot(x, y, kind='reg')
```

```python
# 残差

sns.jointplot(x, y, kind='resid')
```

```python
# 核密度


sns.jointplot(x, y, kind='kde')
```

## 皮尔森相关系数

两个变量之间的相关系数越高，从一个变量去预测另一个变量的精确度就越高。皮尔森相关系数是一种最简单的反应特征和相应之间关系的方法，它衡量的是变量之间的线性相关性，只对线性关系敏感

使用的统计函数是$sci.pearsonr$，除了皮尔森系数外，还返回`p-value`  
`p-value`粗略地表示从一个不相关系统采样所得的数据集产生这样的皮尔森相关系数的概率值

通常情况下，通过以下取值范围判断变量的相关强度

相关系数 | 相关强度
--- | ---
0.8-1.0 | 极强相关
0.6-0.8 | 强相关
0.4-0.6 | 中等程度相关
0.2-0.4 | 弱相关
0.0-0.2 | 极弱相关或无相关

```python
sns.jointplot(x, y, stat_func=sci.pearsonr)
```

```python
tips = sns.load_dataset('tips')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', stat_func=sci.pearsonr)
```
