## FacetGrid

`FacetGrid` 可以构建条件关系的多图网格，很好地提取有关复杂数据的大量信息，直接将绘图结构连接到数据集的结构  
想要使用这些功能，数据必须位于pandas的`DataFrame`中，并且必须采用整洁的数据形式，即：每一行都是一个观测值，每一列都是一个变量  
有一些函数，比如`lmplot`, `catplot`, `pairplot`, `relpolt`等就使用了`FacetGrid`类，不需要过多设置时，直接调用这些函数会更加方便

$FacetGrid$ 可以在三个维度上绘制：$row$, $col$, $hue$  
`Axes`的行和列为两个维度，色调关系视为第三个维度：不同的子集使用不同的颜色来绘制

## 简单的绘制

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='white')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### 初始化网格，设置matplotlib的figure和axes，但不绘制内容

```python
g = sns.FacetGrid(tips, col='time')
```

### 在建立的网格上绘制直方图

```python
g = sns.FacetGrid(tips, col='time')
g.map(plt.hist, 'tip')
```

```python
g = sns.FacetGrid(tips, col='time', hue='smoker')
g.map(plt.scatter, 'total_bill', 'tip', alpha=0.7)

# 添加图例
g.add_legend()
```
