折线图和散点图是非常常用的表现两个变量之间关系的图形。  
+ 散点图是数据点在直角坐标系上的分布图，表示因变量随自变量变化的大致趋势，是进一步分析/拟合的基础
+ 折线图是可视化时间序列的好帮手，能非常好的表现变化趋势

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sns.set()
```

```python
tips = sns.load_dataset('tips')
tips
```

## 散点图

```python
sns.scatterplot(x='total_bill', y='tip', hue='day', style='time', data=tips)
```

```python
# 使用连续颜色显示定义变量
sns.scatterplot(x='total_bill', y='tip', hue='size', size='size', data=tips)
```

```python
# 使用不同颜色映射
sns.scatterplot(x='total_bill', y='tip', hue='size', size='size', palette='cool', data=tips)

# palette 可以是 name, list或者dict, 用来表达不同level的hue的变化
```

```python
# 通过点的大小显示定量变量，限制点的大小范围
sns.scatterplot(x='total_bill', y='tip', hue='size', size='size', palette='cool', sizes=(20, 200), data=tips)
```

## 折线图

```python
# lineplot 是轴级函数，不能直接调整宽高比

plt.figure(figsize=(30, 6))
sns.lineplot(x='total_bill', y='tip', data=tips)
```

如果一个`x`对应多个`y`, seaborn的默认行为是在每个`x`值处聚合多个`y`值，绘制`y`的均值和均值周围的`95%`置信区间

置信区间是使用`bootstrapping`计算的，如果禁用置信区间，绘图的速度会大大加快。适用于时间密集较大的数据集

```python
plt.figure(figsize=(30, 6))
sns.lineplot(x='total_bill', y='tip', data=tips, ci=None)
```

对较大数据的另一个选择是：通过绘制标准差而不是置信区间来表示分布在每个`x`轴上的分布范围

```python
plt.figure(figsize=(24, 6))
sns.lineplot(x='total_bill', y='tip', data=tips, ci='sd')
```

## relplot

在`scatter`和`line`的基础上添加`kind=line`; 默认是`scatter`

$relplot$ 是`figure`级函数，可以直接调整图片的高和宽高比

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='line', height=6, aspect=5)
```
