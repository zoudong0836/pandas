```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

如果想要可视化一个分类属性与连续属性之间的关系，会发现由于分类属性的值有限，绘制的点会发生重叠

```python
sns.scatterplot(x='day', y='tip', data=tips)
```

如果让每个点在水平(或垂直)方向做一定的偏移，就可以更好地展示两者之间的关系了   
`seaborn` 中有针对分类属性散点图的函数 $stripplot$

## 随机抖动散点图

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot('day', 'total_bill', hue='smoker', data=tips, jitter=0.2, dodge=True,
             palette='autumn', size=15, marker='d', edgecolor='gray', alpha=.25)

# 可以通过 jitter 指定抖动的宽度
# 参数 dodge 可以使吸烟者和不吸烟者的散点发生分离
```

## swarm 散点图
> 将点的随机抖动改为沿水平方向依次排列以避免重叠

```python
sns.swarmplot(x='time', y='tip', data=tips, palette='summer', hue='smoker')
```

分类散点图并不常用，在关心一个分类属性的分布情况时，更多选择箱线图和小提琴图

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='cool', inner=None)
sns.swarmplot(x='day', y='total_bill', data=tips, palette='winter', edgecolor='white', alpha=1)
```
