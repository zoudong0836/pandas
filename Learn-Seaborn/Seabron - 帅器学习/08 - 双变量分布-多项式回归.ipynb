当回归函数不能用直线来描述时，要考虑用非线性回归函数。  
多项式回归属于非线性回归的一种，这里指单因子多项式回归(即一元多项式回归)

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
```

```python
ans = sns.load_dataset('anscombe')
ans.head()
```

参数说明

+ $ci:$    int[0, 100] 或 None, 回归估计的置信区间大小，绘制回归线周围的半透明带，对于大型数据集，建议设置为None以避免计算   
+ $order:$   如果`order > 1`, 用`numpy.plotfit`估计多项式回归，多项式次数

```python
sns.regplot(x='x', y='y', data=ans.loc[ans.dataset == 'II'], scatter_kws={"s" : 80}, order=1, ci=None)
```

```python
# 残差图

sns.residplot(x='x', y='y', data=ans.loc[ans.dataset == 'II'])
```

```python
plt.figure(figsize=(6, 6))
sns.regplot(x='x', y='y', data=ans.loc[ans.dataset == 'II'], scatter_kws={"s" : 80}, order=2, ci=None)
```

```python
plt.figure(figsize=(6, 6))
sns.residplot(x='x', y='y', data=ans.loc[ans.dataset == 'II'], order=2, lowess=True)
```
