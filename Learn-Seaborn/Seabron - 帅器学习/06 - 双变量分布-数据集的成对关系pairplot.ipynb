## 用 $pairplot$ 绘制数据集变量的成对关系

```python
import seaborn as sns
sns.set()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

### 绘制每对变量关系的散点图和每个变量自己的直方图

```python
sns.pairplot(iris)
```

### 对单变量图使用核密度估计

```python
sns.pairplot(iris, diag_kind='kde')
```

### 通过绘图元素的颜色显示不同的类别

```python
sns.pairplot(iris, hue='species', palette='Set2')
```

### 将线性回归模型拟合到散点图

```python
sns.pairplot(iris, kind='reg')
```
