set_style() -- 更改 Axes 的颜色，是否启用网格和刻度

5个默认主题
+ darkgrid
+ whitegrid
+ dark
+ white
+ ticks

```python
import seaborn as sns
import numpy as np

tips = sns.load_dataset('tips')
x = tips.tip
x
```

```python
# 默认 
sns.set()
sns.distplot(x, kde=False)
```

```python
sns.set(style='darkgrid')
sns.distplot(x, kde=False)
```

```python
sns.set(style='darkgrid')
sns.distplot(x, kde=True)
```

```python
sns.set(style='whitegrid')
sns.distplot(x, kde=False)
```

```python
sns.set(style='whitegrid')
sns.distplot(x, kde=True)
```

```python
sns.set(style='dark')
sns.distplot(x, kde=False)
```

```python
sns.set(style='white')
sns.distplot(x, kde=False)
```

```python
sns.set(style='ticks')
sns.distplot(x, kde=False)
```

```python
sns.set(style='ticks')
sns.distplot(x, kde=True)
```
