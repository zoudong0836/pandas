displot 函数用于绘制直方图
> 它可以灵活地绘制单变量分布的观测图  
> 它将 matplotlib 中的 hist 函数 (自动计算 bin 大小的直方图) 与 Seaborn kdeplot() 和 rugplot() 函数相结合

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='white')
````

````python
# 生成测试数据

import numpy as np
x1 = np.random.normal(0, 1, 50)
x2 = np.random.normal(5, 3, 50)
x = np.append(x1, x2)
x
````

## 将数组中的数据点绘制为轴上的棒 - rugplot

````python
sns.rugplot(x, height=0.5, color='r')
````

## 核密度函数 kdeplot 的绘制过程

用平滑的峰值函数 $K_h()$ 来拟合观察到的数据点，从而对真实的概率分布曲线进行模拟。$K$为核函数，满足概率密度函数性质，$h$为平滑参数，称为带宽 bandwidth.  
将每个数据点的数据$+$带宽当作核函数的参数，得到`N`个核函数，再线性叠加就形成了估计函数，归一化后就是核密度估计函数了

seaborn 中使用的是高斯核

## kedplot 提供拟合并绘制单变量或双变量核密度估计值

````python
# 在密度曲线下着色并使用非默认颜色

sns.kdeplot(x, shade=True, color='g')
````

+ 带宽反映了$KDE$曲线整体的平坦程度，也即观察到的数据点在$KDE$曲线形成过程中所占的比重 
+ 带宽越大，观察到的数据点在最终形成的曲线形状中所占比重越小，$KDE$整体曲线就越平坦  
+ 带宽越小，观察到的数据点在最终形成的曲线形状中所占比重越大，$KDE$整体曲线就越陡峭  

````python
sns.kdeplot(x, bw=0.05)
````

````python
sns.kdeplot(x, bw=5)
````

## 绘制双变量密度曲线

````python
sns.kdeplot(x1, x2)
````

````python
sns.kdeplot(x1, x2, shade=True)
````

````python
sns.kdeplot(x1, x2, n_levels=10, cmap='Purples_d', cbar=True)
````

##  用 distplot 绘制变量分布关系

```python
distplot(a, bins=None, hist=True, kde=True, rug=False, fit=None, hist_kws=None, kde_kws=None, rug_kws=None, fit_kws=None, color=None, vertical=False, norm_hist=False, axlabel=None, label=None, ax=None)
```
参数|类型|说明
:---|:---|:---
a|Series,1d-array,list|观察数据
bins|matplotlib hist()的参数，可选|无，使用Freedman-Diaconis规则
hist|bool,optional|是否绘制(标准化)直方图
kde|bool,optional|是否绘制高斯核密度估计
rug|bool,optional|是否在支撑轴上绘制rugplot


````python
sns.distplot(x, vertical=True)
````

传入 rugplot、kedplot 和 hist 的参数进行更多设置

````python
sns.distplot(x, rug=True, rug_kws={"color" : "g"}, 
            kde_kws={"color" : "k", "lw" : 3, "label" : "KDE"},
            hist_kws={"histtype" : "step", "linewidth" : 3, "alpha" : 0.5, "color" : "g"})
````
